# 结果汇总

### 提取每个csv文件的前20行然后汇总


```python
columns=['address', 'blockNum', 'transactionHash','positionOriginal','positionEdited', 'ETHER', 'WETH', 'USDT','USDC', 'WBTC', 'DAI']
df=pd.DataFrame(columns=columns)
df
```

```python
nameList=["addrBalanceDiff_specialTokens_sortedByEther","addrBalanceDiff_specialTokens_sortedByWeth","addrBalanceDiff_specialTokens_sortedByUsdt","addrBalanceDiff_specialTokens_sortedByUsdc","addrBalanceDiff_specialTokens_sortedByWbtc","addrBalanceDiff_specialTokens_sortedByDai"]
```

```python
for name in nameList:
    path="../data/csv/"+name+".csv"
    tempDf = pd.read_csv(path)
    df=df.append(tempDf.iloc[0:20])
```

```python
temp_list=np.array(range(1,len(df)+1,1))
df.index=pd.Series(temp_list)
```

```python
df.to_csv("../data/csv/addrBalanceDiff_specialTokens_part.csv")
```
