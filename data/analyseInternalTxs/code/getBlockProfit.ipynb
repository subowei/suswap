```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv('../data/csv/addrBalanceDifference.csv')
```

```python
block_balance={}
```

```python
for index, row in df.iterrows():
    blockNum=row["blockNum"]
    balanceDifference=row["balanceDifference"]
    if float(balanceDifference)==0:continue
    if block_balance.get(blockNum)!=None:
        block_balance_value=block_balance[blockNum]
        block_balance_value['balanceDifference']+=balanceDifference
        block_balance[blockNum]=block_balance_value
    else:
        block_balance_value={"balanceDifference":balanceDifference,"serviceChargeDiff":row["serviceChargeDiff"]}
        block_balance[blockNum]=block_balance_value
```

```python
df=pd.DataFrame(columns=('blockNum','balance_service_Difference','balanceDifference','serviceChargeDiff'))
for key,value in block_balance.items():
    balance_service_Difference=float(value["balanceDifference"]+value["serviceChargeDiff"])
    balanceDifference=float(value["balanceDifference"])
    serviceChargeDiff=float(value["serviceChargeDiff"])
                                     
    df=df.append([{'blockNum':key,'balance_service_Difference':balance_service_Difference,'balanceDifference':balanceDifference,'serviceChargeDiff':serviceChargeDiff}],ignore_index = True)
```

```python
df1=df.sort_values("balance_service_Difference",ascending=False)
temp_list=np.array(range(1,len(df)+1,1))
df1.index=pd.Series(temp_list)
df1
```

```python
df1.to_csv("../data/csv/blockProfit1.csv")
```
